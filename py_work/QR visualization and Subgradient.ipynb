## subgradient G(beta)_beta_j

```python
# subgradient QR_beta_j
def QR_subgradient(res, j):
    if res > 0:
        return -(tau/n)*sum(X[:,j].reshape(-1,1))
    elif res < 0:
        return -((tau-1)/n)*sum(X[:,j].reshape(-1,1))
    elif res == 0:
        return -(random.uniform(tau-1, tau)/n)*sum(X[:,j].reshape(-1,1))
```

```python
# subgradient G_beta_j
def G_beta_j(res, j):
    if beta[j] > 0:
        QR_subgradient(res, j) + ld
    elif beta[j] < 0:
        QR_subgradient(res, j) - ld
    elif beta[j] == 0:
        QR_subgradient(res, j) + random.uniform(-ld, ld)
```
